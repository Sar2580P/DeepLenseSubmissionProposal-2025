```python
import os
import matplotlib.pyplot as plt
import numpy as np
import glob

os.chdir('/home/sarvagya/DeepLenseSubmissionProposal-2025')
```

```python
# load random 5 files from the directory
# plot these files
N = 7
files = glob.glob('data/Samples/*.npy')
print(f"Total files: {len(files)}")
files = np.random.choice(files, N)

fig, axs = plt.subplots(1, N, figsize=(20, 4))
for i, file in enumerate(files):
    data = np.load(file).T
    print(data.shape)
    axs[i].imshow(data, cmap='gray')
    axs[i].axis('off')
    axs[i].set_title(file.split('/')[-1])
plt.show()
```
